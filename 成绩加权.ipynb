# 成绩加权程序设计

## 1.随机生成一些成绩，注意成绩的合理性
    语文
    数学
    英语
## 2.设定权重参数
    平时成绩占比w1
    期中成绩占比w2
    期末成绩占比w3
    w1+w2+w3=1
## 3.判别器设计

```python
import numpy as np
import random
```

```python
w1 = float(input("请输入平时成绩的占比： "))
w2 = float(input("请输入中期成绩占比： "))
w3 = float(input("请输入期末成绩占比： "))
```

```python
weight_array = np.array([w1, w2, w3])
```

```python
weight_array
```

```python
'''随机生成成绩[语文， 数学， 英语]'''
def generate_grade(students=100, mu=60, sigma=10): # 这里思考一个问题，mu和sigma不合理会怎么样？
    grade_list = np.floor(np.random.normal(mu, sigma,size=(students,3)))
    return grade_list
```

```python
def discriminator(grades):
    grades = np.array(grades)
    final_grade = np.dot(weight_array, grades)
    return final_grade
```

```python
students = 100
grade_list = generate_grade(students)
with open('assess_student.csv', 'w', encoding="utf-8") as f:
    for grades in grade_list:
        grades = list(grades)
        final_grade = float(discriminator(grades))
        if final_grade >= 80:
            assessment = "优"
        elif final_grade >= 70:
            assessment = "良"
        elif final_grade >= 60:
            assessment = "中"
        else:
            assessment = "差"
        grades.append(assessment)
        grades = [str(grade) for grade in grades]
        print(grades)
        grades = ','.join(grades)
        f.write(grades+'\n')
```
