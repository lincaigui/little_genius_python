# 成绩加权程序设计

## 1.随机生成一些成绩，注意成绩的合理性
    平时成绩
    期中成绩
    期末成绩
## 2.设定权重参数
    平时成绩占比w1
    期中成绩占比w2
    期末成绩占比w3
    w1+w2+w3=1
## 3.判别器设计

```python
import numpy as np #numpy是科学计算包，可以进行矩阵，向量的计算
import random
```

## 评定系数，可以n个，注意:
## w1 + w2 + w3 .....+ wn = 1

```python
w1 = float(input("请输入平时成绩的占比： "))
w2 = float(input("请输入中期成绩占比： "))
w3 = float(input("请输入期末成绩占比： "))
```

```python
weight_array = np.array([w1, w2, w3])
```

```python
n_w = len(list(weight_array))
```

```python
'''随机生成成绩[平时成绩， 期中成绩， 期末成绩]'''
def generate_grade(students=100, mu=60, sigma=10, n=3): # 这里思考一个问题，mu和sigma不合理会怎么样？模拟数据不合理可能会有超过100或者低于0，如何处理
    grade_list = np.floor(np.random.normal(mu, sigma,size=(students, n)))
    return grade_list
```

```python
def discriminator(grades):
    final_grade = np.dot(weight_array, grades)
    return final_grade
```

```python
students = 100 #100个学生
grade_list = generate_grade(students, n=n_w)
with open('assess_student.csv', 'w', encoding="utf-8") as f:
    for grades in grade_list:   
        final_grade = discriminator(grades)
        grades = list(grades)
        if final_grade >= 80:
            assessment = "优"
        elif final_grade >= 70:
            assessment = "良"
        elif final_grade >= 60:
            assessment = "中"
        else:
            assessment = "差"
        grades.append(assessment)
        grades = [str(grade) for grade in grades]
        print(grades)
        grades = ','.join(grades)
        f.write(grades+'\n')
```
